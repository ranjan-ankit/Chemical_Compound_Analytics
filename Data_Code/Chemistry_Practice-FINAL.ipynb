#### Predict ExactMolWt(m), MolLogP(m), CalcTPSA(m)

```python
from rdkit.Chem.Descriptors import ExactMolWt
from rdkit.Chem.Crippen import MolLogP
from rdkit.Chem.rdMolDescriptors import CalcTPSA
from rdkit import Chem

MolWt=[]
MolLogp=[]
CalTPSA=[]
for i in y_temp['SMILES']:
    m=Chem.MolFromSmiles(i)
    MolWt.append(ExactMolWt(m))
    MolLogp.append(MolLogP(m))
    CalTPSA.append(CalcTPSA(m))
    
y_temp['Mol_Wt']=MolWt
y_temp['Mol_log_P']=MolLogp
y_temp['Calc_TPSA']=CalTPSA
y_temp
```

```python
from rdkit.Chem.AtomPairs import Pairs

ms = [Chem.MolFromSmiles(i) for i in y_temp['SMILES']]
pair_vec = [Pairs.GetAtomPairFingerprintAsBitVect(x) for x in ms]
pair_vect_binary = [i.ToBitString() for i in pair_vec]
```
